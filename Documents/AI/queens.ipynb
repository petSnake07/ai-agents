## Step 1: Create the CSP Object
- Implement the function _queens_constraint_ that will be used to check the arc consistency between two variables connected by a binary constraint.
- Implement the function _build_csp_ to create a CSP object representing the puzzle.  

Take a look at the [csp.ipynb](./csp.ipynb) notebook, to figure out what you need to specify to instantiate the CSP object.

```python
def queens_constraint(var1, val1, var2, val2):
    """
    This function takes two neighboring variables (variables connected 
    by a binary constraint) along with one value from each of their domains.
    It returns True if assigning val1 to var1 and val2 to var2 satisfies the
    queens constraint (no shared row or shared diagonal).
    :param var1 (int): the first variable (first queen's column)
    :param val1 (int): value from the domain of the variable var1 (first queens's row)
    :param var2 (int): the second variable (second queen's column)
    :param val2 (int): value from the domain of the variable var2 (second queens's row)
    :return: (boolean) True if assignment is consistent, False
        otherwise
    """
    # Enter your code here and remove the pass statement below
    return val1 != val2 and abs(val1 - val2) != abs(var1 - var2)


def build_csp(n, q1):
    """
    Create a CSP object representing the puzzle.
    :param n: the size of the board (number of queens)
    :param q1: the row position associated with the queen in column 1
    :return: CSP object
    """
    # create the neighbors dictionary
    # create the domains dictionary (make sure you apply node consistency)
    # use all the above to create the CSP object
    # Enter your code here and remove the pass statement below
    neighbors = {}
    for var in range(1, n + 1):
        neighbors[var] = set()
        for other in range(1, n + 1):
            if other != var:
                neighbors[var].add(other)

    domains = {}
    for var in range(1, n + 1):
        if var == 1:
            domains[var] = {q1}
        else:
            domains[var] = set(range(1, n + 1))

    return CSP(domains, neighbors, queens_constraint) 
```

## Step 2:  Backtracking Search
Implement the function _btrack_. 
    
This function takes two integer arguments:  the size of the board (number of queens) and the row associated with Queen 1.  

The function should call build_csp to create the CSP object then invoke the basic backtracking search algorithm on that CSP object.

The function returns a **tuple**: the solution (a dictionary representing a complete and valid assignment), and the CSP object.


```python
def btrack(n, q1):
    """
    Solve the given puzzle with basic backtracking search
    :param n: the size of the board (number of queens)
    :param q1: the row position associated with the queen in column 1
    :return: a tuple consisting of a solution (dictionary) and the
             CSP object.
    """
    # Enter your code here and remove the pass statement below
    csp = build_csp(n,q1)
    solution = csp.backtracking_search()
    return solution,csp
```

## Step 3:  Backtracking Search with AC-3
Implement the function _btrackac3_.

This function takes two integer arguments: the size of the board (number of queens) and the row associated with Queen 1.

The function should still call _build_csp_ to create the CSP object.

It will run the AC-3 algorithm as a preprocessing step on the CSP object before invoking the basic backtracking search algorithm .

The function returns a tuple: the solution (a dictionary representing a complete and valid assignment), and the CSP object.

```python
def btrackac3(n, q1):
    """
    Solve the given puzzle with backtracking search and  AC-3 as a 
    preprocessing step.
    :param n: the size of the board (number of queens)
    :param q1: the row position associated with the queen in column 1
    :return: a tuple consisting of a solution (dictionary) and the
             CSP object.
    """
    # Enter your code here and remove the pass statement below
    csp = build_csp(n,q1)
    csp.ac3_algorithm()
    solution = csp.backtracking_search()
    return solution, csp
```

With this implementation you should be able to find a solution for these puzzles more efficiently.  The AC3 algorithm reduces the domain of all the variables before invoking the backtracking search.
